Para este proyecto se ha utilizado un dataset almacenado en kaggle, el cual contiene información de caracteristicas de laptops ofertadas en la página de tecnología souq.com.

    - Marca de portátil : contiene el nombre de la marca
    - Sistema operativo : contiene el tipo de sistema operativo
    - HDC : contiene la capacidad del disco duro
    - Processor_Family : contiene el tipo de familia de procesos
    - Tamaño de pantalla : contiene el tamaño de pantalla de cada computadora portátil
    - Tamaño de la memoria : contiene el tamaño de la memoria de cada computadora portátil
    - Precio : Contiene el valor de venta para cada portátil
    - Descripción : contiene la descripción de cada computadora portátil
    - URL : contiene el enlace de la página


El archivo contiene 2280 productos y se encuentra disponible en el siguiente [link](https://www.kaggle.com/najwaalharbi/souqcom-laptops-dataset)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

path = 'souq_laptops.csv'
col = ['laptop brand','Operating system','HDC','Processor_Family','Screen_Size','Memory_Size','price','Description','URL']
df = pd.read_csv(path, usecols = col)
```

1) Agrupación de laptop de acuerdo a su marca

```python
marca = df['laptop brand'].value_counts()
grafico1 = plt.figure(figsize=(8,8))
plt.pie(marca,labels=marca.index)
plt.axis('equal')
plt.show()
```

2) Porcentaje de laptops de acuerdo al sistema operativo que utilizan

```python
so = df['Operating system'].value_counts()
grafico2 = plt.figure(figsize=(8,8))
plt.pie(so, labels = so.index, autopct = '%1.1f%%')
plt.axis('equal')
plt.show()
```

3) Porcentaje de laptops de acuerdo a su memoria ram

```python
ram = df['Memory_Size'].value_counts()
grafico3 = plt.figure(figsize=(8,8))
plt.pie(ram, labels = ram.index, autopct = '%1.1f%%')
plt.axis('equal')
plt.show()
```

4) Porcentaje de laptops de acuerdo al tamano de su pantalla

```python
hd = df['HDC'].value_counts()
grafico4 = plt.figure(figsize=(8,8))
plt.pie(hd, labels = hd.index, autopct = '%1.1f%%')
plt.axis('equal')
plt.show()
```

5) Los 10 procesadores mas usados

```python
proc = df['Processor_Family'].value_counts().head(10)
grafico5 = plt.figure(figsize=(8,8))
plt.pie(proc, labels = proc.index, autopct = '%1.1f%%')
plt.axis('equal')
plt.show()
```

6) Laptops Acer agrupadas por procesador

```python
acer = df.loc[df['laptop brand']=='Acer'].groupby(by='Processor_Family')['Processor_Family']
grafico6 = plt.figure(figsize=(8,8))
plt.bar(acer.count().index, acer.count().values)
plt.xticks(rotation=90)
plt.show()
```

7) Laptops Acer agrupadas por sistema operativo

```python
acerSo = df.loc[df['laptop brand']=='Acer'].groupby(by='Operating system')['Operating system']
grafico7 = plt.figure(figsize=(8,8))
plt.bar(acerSo.count().index, acerSo.count().values)
plt.xticks(rotation=90)
plt.show()
```

8) Laptops agrupadas por intervalo de precios

```python
precio = pd.cut(pd.to_numeric(df['price']), bins=[500, 1000, 1500, 2000, 2500])
grafico8 = precio.value_counts(sort=False).plot.bar(rot=0, figsize=(10,5))
plt.show()
```

9) Clasifica a las laptops de precio mayor a 4500AED

```python
precioM = pd.to_numeric(df['price']>4500)
grafico9 = precioM.value_counts().plot.bar(rot=0, figsize=(10,5))
plt.show()
```

```python
10) Laptops agrupadas por intervalo de capacidad del disco duro
```

```python
disco = pd.cut(pd.to_numeric(df['HDC']), bins=[0, 1,  256, 512, 750, 1000])
grafico10 = disco.value_counts(sort=False).plot.bar(rot=0, figsize=(10,5))
plt.show()
```
